```python
import numpy as np
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

def power_utility(r, x):
    if r > 0:
        return x ** r
    elif r == 0:
        return np.log(x)
    elif r < 0:
        return -x ** r

r = 0
x_values = np.linspace(0, 100)
y_values = power_utility(r, x_values)
plt.plot(x_values, y_values)
```
